## Import dan Load

Import library yang diperlukan

```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, cross_val_score

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams

import joblib
```

Mengatur Style

```python
rcParams['figure.figsize'] = 15, 10
mpl.style.use(['ggplot'])
```

Load data train

```python
df_train = pd.read_csv('https://drive.google.com/uc?id=1MiU_0V6-LvZ_CyCuWf2HoW-WXNKSIFJQ')
```

Load hasil seleksi fitur

```python
rf1_features = joblib.load('rf1_features.pkl')
rf2_features = joblib.load('rf2_features.pkl')
rf3_features = joblib.load('rf3_features.pkl')
```

Df_train dengan fitur 1

```python
df_train1 = df_train[rf1_features]
df_train1.head()
```

Df_train dengan fitur 2

```python
df_train2 = df_train[rf2_features]
df_train2.head()
```

Df_train dengan fitur 3

```python
df_train3 = df_train[rf3_features]
df_train3.head()
```

## X_train dan y_train

```python
X1_train = df_train1.iloc[:,:-1]
X1_train.head()
```

```python
y1_train = df_train1.iloc[:,-1]
y1_train.head()
```

```python
X2_train = df_train2.iloc[:,:-1]
X2_train.head()
```

```python
y2_train = df_train2.iloc[:,-1]
y2_train.head()
```

```python
X3_train = df_train3.iloc[:,:-1]
X3_train.head()
```

```python
y3_train = df_train3.iloc[:,-1]
y3_train.head()
```

## Model tanpa Tuning

RF1

```python
rf1_non_tuning = RandomForestClassifier()
rf1_non_tuning_acc = cross_val_score(rf1_non_tuning, X1_train, y1_train, scoring="accuracy", cv=10)
print("Akurasi masing-masing = ", rf1_non_tuning_acc)
print("Rata-rata = ", rf1_non_tuning_acc.mean())
```

RF2

```python
rf2_non_tuning = RandomForestClassifier()
rf2_non_tuning_acc = cross_val_score(rf2_non_tuning, X2_train, y2_train, scoring="accuracy", cv=10)
print("Akurasi masing-masing = ", rf2_non_tuning_acc)
print("Rata-rata = ", rf2_non_tuning_acc.mean())
```

RF3

```python
rf3_non_tuning = RandomForestClassifier()
rf3_non_tuning_acc = cross_val_score(rf3_non_tuning, X3_train, y3_train, scoring="accuracy", cv=10)
print("Akurasi masing-masing = ", rf3_non_tuning_acc)
print("Rata-rata = ", rf3_non_tuning_acc.mean())
```

## Hyperparameter Tuning

### Parameter yang akan dituning

Definisi parameter

```python
parameters_rf = {
        'n_estimators' : [50, 100, 150, 200],
        'max_depth' : [None, 3, 6, 9],
        'min_samples_split' : [2, 3, 4, 5]      
        }
```

### RF1

```python
rf1_tuning = RandomForestClassifier()
```

Gridsearch (sekitar 3 menit running)

```python
search_rf1 = GridSearchCV(rf1_tuning,
                        parameters_rf,
                        scoring='accuracy',
                        cv = 10,
                        verbose=3)
search_rf1.fit(X1_train, y1_train)
```

Akurasi dan parameter terbaik pada Gridsearch

```python
rf1_best_acc_score = search_rf1.best_score_
rf1_best_parameter = search_rf1.best_params_
print("Akurasi terbaik = ", rf1_best_acc_score)
print("Parameter terbaik = ", rf1_best_parameter)
```

Buat model dengan parameter terbaik

```python
max_depth= rf1_best_parameter['max_depth']
min_samples_split = rf1_best_parameter['min_samples_split']
n_estimators = rf1_best_parameter['n_estimators']
rf1_tuning = RandomForestClassifier(max_depth=max_depth, min_samples_split=min_samples_split, n_estimators=n_estimators)
```

Cross Val akhir

```python
rf1_tuning_acc = cross_val_score(rf1_tuning, X1_train, y1_train, scoring="accuracy", cv=10)
print("Akurasi masing-masing = ", rf1_tuning_acc)
print("Rata-rata = ", rf1_tuning_acc.mean())
```

### RF2

```python
rf2_tuning = RandomForestClassifier()
```

Gridsearch (sekitar 3 menit running)

```python
search_rf2 = GridSearchCV(rf2_tuning,
                        parameters_rf,
                        scoring='accuracy',
                        cv = 10,
                        verbose=3)
search_rf2.fit(X2_train, y2_train)
```

Akurasi dan parameter terbaik

```python
rf2_best_acc_score = search_rf2.best_score_
rf2_best_parameter = search_rf2.best_params_
print("Akurasi terbaik = ", rf2_best_acc_score)
print("Parameter terbaik = ", rf2_best_parameter)
```

Membuat model dengan parameter terbaik

```python
max_depth= rf2_best_parameter['max_depth']
min_samples_split = rf2_best_parameter['min_samples_split']
n_estimators = rf2_best_parameter['n_estimators']
rf2_tuning = RandomForestClassifier(max_depth=max_depth, min_samples_split=min_samples_split, n_estimators=n_estimators)
```

Cross val akhir

```python
rf2_tuning_acc = cross_val_score(rf2_tuning, X2_train, y2_train, scoring="accuracy", cv=10)
print("Akurasi masing-masing = ", rf2_tuning_acc)
print("Rata-rata = ", rf2_tuning_acc.mean())
```

### RF3

```python
search_rf3 = GridSearchCV(rf3_tuning,
                        parameters_rf,
                        scoring='accuracy',
                        cv = 10,
                        verbose=3)
search_rf3.fit(X3_train, y3_train)
```

Akurasi dan parameter terbaik dari Gridsearch

## Non Tuning vs Tuning

Model yang dipilih antara tuning atau non-tuning

```python
model_yang_dipilih = ""

if (rf1_tuning_acc.mean() > rf1_non_tuning_acc.mean()):
  rf1_model = rf1_tuning
  model_yang_dipilih += "RF1 Tuning, "
else:
  rf1_model = rf1_non_tuning
  model_yang_dipilih += "RF1 Non Tuning, "

if (rf2_tuning_acc.mean() > rf2_non_tuning_acc.mean()):
  rf2_model = rf2_tuning
  model_yang_dipilih += "RF2 Tuning, "
else:
  rf2_model = rf2_non_tuning
  model_yang_dipilih += "RF2 Non Tuning, "

if (rf3_tuning_acc.mean() > rf3_non_tuning_acc.mean()):
  rf3_model = rf3_tuning
  model_yang_dipilih += "RF3 Tuning"
else:
  rf3_model = rf3_non_tuning
  model_yang_dipilih += "RF3 Non Tuning"

print("Model yang dipilih = ", model_yang_dipilih)
```

Visualisasi perbandingan tuning vs non-tuning

```python
def addlabels(x,y,z):
    for i in range(len(x)):
      plt.text(i, x[i]+0.02, format(x[i], ".3f"), ha = 'center')
      plt.text(i+0.25, y[i]+0.02, format(y[i], ".3f"), ha = 'center')
      plt.text(i+0.5, z[i]+0.02, format(z[i], ".3f"), ha = 'center')
```

```python
N = 3
ind = np.arange(N) 
width = 0.25

plt.figure(figsize=(8,9))
non_tuning_acc = [rf1_non_tuning_acc.mean(),  rf2_non_tuning_acc.mean(),  rf3_non_tuning_acc.mean()]
bar1 = plt.bar(ind, non_tuning_acc, width, color='goldenrod')
  
tuning_acc = [rf1_tuning_acc.mean(),  rf2_tuning_acc.mean(),  rf3_tuning_acc.mean()]
bar2 = plt.bar(ind+width, tuning_acc, width, color='forestgreen')

selisih = []
for i in range(0, len(non_tuning_acc)):
  selisih.append(np.abs(non_tuning_acc[i]-tuning_acc[i]))

bar3 = plt.bar(ind+width*2, selisih, width, color = 'firebrick')

addlabels(non_tuning_acc, tuning_acc, selisih)
plt.xlabel("Model", fontweight='bold')
plt.ylabel('Accuracy', fontweight='bold')
plt.ylim([0,0.9])

plt.xticks(ind+width,['RF1', 'RF2', 'RF3'])
plt.legend( (bar1, bar2, bar3), ('Non Tuned', 'Tuned', 'Difference') )
plt.show()
```
